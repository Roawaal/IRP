```python
import pandas as pd

# Function to process each CSV file
def process_csv(file_name):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_name)

    # Sort the DataFrame based on the "Import (US$ Thousand)" column in descending order
    sorted_df = df.sort_values(by='Import (US$ Thousand)', ascending=False)

    # Extracting the top one row from the sorted DataFrame
    top_one = sorted_df.head(1)

    # Selecting only the "Partner Name", "Year", and "Import (US$ Thousand)" columns
    selected_columns = top_one[['Partner Name', 'Year', 'Import (US$ Thousand)']]

    # Printing the selected columns
    print(f"Top Import for {file_name}:")
    print(selected_columns)

    # Filter rows based on specified countries
    countries_of_interest = ['World', 'Europe & Central Asia', 'East Asia & Pacific', 
                             'North America', 'Middle East & North Africa', 
                             'Latin America & Caribbean', 'Sub-Saharan Africa', 
                             'South Asia']
    filtered_df = df[df['Partner Name'].isin(countries_of_interest)]

    selected_columns_filtered = filtered_df[['Partner Name', 'Import (US$ Thousand)']]

    # Printing the filtered columns
    print(f"\nFiltered Imports for {file_name}:")
    print(selected_columns_filtered)

# Process 2019.csv
process_csv('trade/2019.csv')

# Process 2020.csv
process_csv('trade/2020.csv')

# Process 2021.csv
process_csv('trade/2021.csv')
```
